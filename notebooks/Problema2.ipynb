### 2. ¬øPor qu√© las componentes By y vy permanecen aproximadamente cero?

**Respuesta:**
En una onda de Alfv√©n pura propag√°ndose a lo largo del eje z, las perturbaciones ocurren solo en las direcciones perpendiculares al campo magn√©tico principal. Como el campo inicial es B‚ÇÄ√™_z y la perturbaci√≥n es B‚ÇÅ√™_x, no hay componentes en la direcci√≥n y, por lo que By ‚âà 0 y vy ‚âà 0.

El an√°lisis confirma que estas componentes permanecen en cero durante toda la simulaci√≥n.


```python
# Mostrar verificaci√≥n de By y vy
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img_by_vy_path = '../results/problema2/by_vy_t10.png'
if os.path.exists(img_by_vy_path):
    img_by_vy = mpimg.imread(img_by_vy_path)
    plt.figure(figsize=(10, 6))
    plt.imshow(img_by_vy)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
else:
    print("‚ùå Imagen de verificaci√≥n By/vy no encontrada.")
```

### 4. ¬øEn qu√© r√©gimen se encuentra la simulaci√≥n: sub-Alfv√©nico o s√∫per-Alfv√©nico?

**Respuesta:**
El r√©gimen se determina comparando la velocidad de perturbaci√≥n con la velocidad de Alfv√©n. Si la velocidad de perturbaci√≥n es menor que v_A, el r√©gimen es sub-Alfv√©nico (M_A < 1); si es mayor, es s√∫per-Alfv√©nico (M_A ‚â• 1).

Seg√∫n los resultados, la simulaci√≥n se encuentra en r√©gimen **sub-Alfv√©nico** (M_A < 1), lo que significa que las perturbaciones se propagan m√°s lentamente que la velocidad de Alfv√©n.

### 5. ¬øC√≥mo evoluciona la amplitud de la perturbaci√≥n en el tiempo?
